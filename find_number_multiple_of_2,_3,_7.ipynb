find out the number multiple of 2, 3, 7 from 1 to 100,000

```python
count_2 = 0
count_3 = 0
count_7 = 0

for i in range(1, 100001):
    if i % 2 == 0:
        count_2 += 1
    if i % 3 == 0:
        count_3 += 1
    if i % 7 == 0:
        count_7 += 1

print("Multiples of 2:", count_2)
print("Multiples of 3:", count_3)
print("Multiples of 7:", count_7)
```
